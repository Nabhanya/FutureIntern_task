```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
sns.set_style("darkgrid")
data= pd.read_csv("Iris.csv")
data.head()
```

```python
#Scatter plot with 'SepalLengthCm' on the x-axis and 'SepalWidthCm' on y-axis
sns.scatterplot(x="SepalLengthCm",y="SepalWidthCm",data=data,hue='Species')
plt.show()
```

```python
#Scatter plot with 'PetalLengthCm' on the x-axis and 'PetalWidthCm' on y-axis
sns.scatterplot(x="PetalLengthCm",y="PetalWidthCm",data=data,hue='Species')
plt.show()
```

```python
for col in data.columns[ :-1]:
    sns.boxplot(x="Species",y=col,data=data)
    plt.show()
```

```python
for col in data.columns[:-1]:
    sns.histplot(x=col,data=data)
    plt.show()
```
